```python
# List for holding lat_lngs and cities
#lat_lngs = []
#cities = []

# Create a set of random lat and lng combinations
#lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
#lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
#lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
#for lat_lng in lat_lngs:
#    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
#    if city not in cities:
#        cities.append(city)

# Print the city count to confirm sufficient count
#len(cities)


# I didn't like the given method, as quite frankly it doesn't guarentee over 500 cities.
# If our random sample doesn't give us 500 unique cities, then sure we print out the city
# count, but that doesn't magically make more cities appear. What are we gonna do, rerun our code?
# No. In fact we barely cleared 500 after we put the working cities in our array.
# Here's a way that guarentees more than 500 cities. We need to not only have more than 500
# cities in our cities array, but also for enough of them to have sufficient data to make it to
# our dataframe. So we cut out the middle man, and use a while loop that won't stop until we have enough.


# We prepare our url
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
# And our data frame
city_data=pd.DataFrame()
city_data['City']=""
city_data['Cloudiness']=""
city_data['Country']=""
city_data['Date']=""
city_data['Humidity']=""
city_data['Lat']=""
city_data['Lng']=""
city_data['Max Temp']=""
city_data['Wind Speed']=""

#we also have some variables to keep track of counts
index=0
record=1
#print some introductory stuff
print('Beginning Data Retrieval')
print('--------------------------------')

#here we have a while loop that won't stop until we have at least 1000 cities for our dataframe
#(500 would work as well but I want more data points for fun)
while city_data['City'].count() < 1000:
    #here we exclude the north pole as np.random.uniform is exclusive on its upper bound
    #hopefully santa won't be tooo mad
    lat = np.random.uniform(low=-90.000, high=90.000)
    lng = np.random.uniform(low=-180.000, high=180.000)
    #we get the city nearest to our coords using citipy
    city = citipy.nearest_city(lat, lng).city_name
    
    #we check to make sure we don't already have the city
    if city not in city_data['City']:
        #construct our url for a call to the openweather api
        query_url = url + "appid=" + api_key + "&q=" + city +"&units=" + units
        weather_response = requests.get(query_url)
        weather_json = weather_response.json()
    
        #now we try to add the city to our dataframe. If no information is missing it will be added
        try:
            print(f"Processing Record {record} | {city}")
            city_data.loc[index, 'City']=city
            city_data.loc[index, 'Cloudiness']=weather_json['clouds']['all']
            city_data.loc[index, 'Country']=weather_json['sys']['country']
            city_data.loc[index, 'Date']=weather_json['dt']
            city_data.loc[index, 'Humidity']=weather_json['main']['humidity']
            city_data.loc[index, 'Lat']=weather_json['coord']['lat']
            city_data.loc[index, 'Lng']=weather_json['coord']['lon']
            city_data.loc[index, 'Max Temp']=weather_json['main']['temp_max']
            city_data.loc[index, 'Wind Speed']=weather_json['wind']['speed']
            #we only increase the index if its all successful
            index+=1
        
        #otherwise we print that we couldn't get the data
        except KeyError:
            print("City not found. Skipping...")
        #increase our count of which city attempt this is
        record+=1
#once we finish, we print an ending message        
print('--------------------------------')
print('Data Retrieval Complete')
print('--------------------------------')
```
